```python
import time
from PIL import Image, ImageDraw

# Import mattplotlib to plot image outputs
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['image.cmap'] = 'gray'
 
```

```python
import os
import subprocess

brew_prefix = subprocess.check_output(["/opt/homebrew/bin/brew", "--prefix"]).decode().strip()
os.environ["DYLD_FALLBACK_LIBRARY_PATH"] = f"{brew_prefix}/lib:" + os.environ.get("DYLD_FALLBACK_LIBRARY_PATH", "")


import torch
from transformers import AutoModelForCausalLM, AutoTokenizer


model = AutoModelForCausalLM.from_pretrained(
    "moondream/moondream-2b-2025-04-14",
    trust_remote_code=True,
    device_map="auto",
)
 
dtype = next(model.parameters()).dtype
print(dtype)
```

```python
# Visual Querying
qimg = Image.open('./test_images/man.png')
query = "Whats going on here?"
print(f"\nVisual query: '{query}'")
print(model.query(qimg, query)["answer"])
```

```python
print('Normal caption:')

t1 = time.time()
normal_caption = model.caption(qimg, length="normal")["caption"]
 
for t in normal_caption:
    print(t, end="", flush=True)
    t2 = time.time()
    diff = t2 - t1
    print(f"\Caption Time : {round(diff,2)}")

print(normal_caption)
```
